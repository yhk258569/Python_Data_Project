```python
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

df['job_skills'] = df['job_skills'].apply(
    lambda x: ast.literal_eval(x) if pd.notna(x) else x
)
```

```python
df = df[df['job_title_short'] == 'Data Analyst'].copy()
```

```python
work_from_home = df['job_work_from_home'].value_counts()

plt.figure(figsize = (6,6))

plt.pie(
    work_from_home,
    labels = ['False', 'True'],
    autopct = '%1.1f%%',
    startangle = 90
)

plt.title('Work from Home Status')
plt.show()
```

```python
job_no_degree = df['job_no_degree_mention'].value_counts()
job_helath_insurance = df['job_health_insurance'].value_counts()

fig, ax = plt.subplots(1, 3)
fig.set_size_inches((12, 5))

plt.subplot(1, 3, 1)
plt.pie(work_from_home, labels = ['False', 'True'], autopct = '%1.1f%%', startangle = 90)
plt.title('Wokr from Home Offered')

plt.subplot(1, 3, 2)
plt.pie(job_no_degree, labels = ['False', 'True'], autopct = '%1.1f%%', startangle = 90)
plt.title('Degree Requirement')

plt.subplot(1, 3, 3)
plt.pie(job_helath_insurance, labels = ['False', 'True'], autopct = '%1.1f%%', startangle = 90)

fig.tight_layout()
plt.show()
```

```python
dict_column = {
    'job_work_from_home': 'Work from Home Offered',
    'job_no_degree_mention': 'Degree Requirement',
    'job_health_insurance': 'Health Insurance Offered'
}

fig, ax = plt.subplots(1, 3)
fig.set_size_inches((12, 5))

for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(
        df[column].value_counts(),
        labels = ['False', 'True'],
        autopct = '%1.1f%%',
        startangle = 90
    )
    ax[i].set_title(title)

plt.show()
```

```python
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

df['job_skills'] = df['job_skills'].apply(
    lambda x: ast.literal_eval(x) if pd.notna(x) else x
)
```

```python
health_insurance = df['job_health_insurance'].value_counts()

plt.figure(figsize = (6,6))
plt.pie(
    health_insurance,
    labels = ['False', 'True'],
    autopct = '%1.1f%%',
    startangle = 90
)

plt.title('Health Insurance Offered')
plt.show()
```

```python
job_schedule = df['job_schedule_type'].value_counts().head(3)

plt.figure(figsize=(6,6))
plt.pie(
    job_schedule,
    labels=job_schedule.index,
    autopct='%1.1f%%',
    startangle=90
)

plt.title('Job Schedule Distribution')
plt.show()
```

```python
degree = df['job_no_degree_mention'].value_counts()

plt.figure(figsize = (6,6))
plt.pie(
    degree,
    labels = ['False', 'True'],
    autopct = '%1.1f%%',
    startangle = 90
)

plt.title('Degree Requirement')
plt.show()
```
