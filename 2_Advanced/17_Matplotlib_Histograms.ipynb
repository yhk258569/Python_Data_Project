```python
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()
```

```python
df_DA_US['salary_year_avg'].plot(kind = 'hist', bins = 40, edgecolor = 'black')
plt.xlim(0, 250000)

ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))

plt.title('Distribution of United States Data Analyst Yearly Salaries')
plt.xlabel('Yearly Salary')
plt.ylabel('Number of Jobs')
plt.show()
```

```python
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DS = df[df['job_title_short'] == 'Data Scientist'].copy()

df_DS = df_DS.dropna(subset=['salary_hour_avg'])

df_DS.plot(
    kind = 'hist',
    bins = 30,
    edgecolor = 'black'
)

plt.title('Data Scientist Hourly Salaries')
plt.xlabel('Hourly Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
df['skill_count'] = df['job_skills'].apply(
    lambda x: len(x) if isinstance(x, list) else 0
)

df['skill_count'].plot(
    kind = 'hist',
    bins = 20,
    edgecolor = 'black'
)

plt.title('Job Postings by Number of Skills')
plt.xlabel('Number of Skills')
plt.ylabel('Number of Jobs')
plt.show()
```

```python
df_DE = df[df['job_title_short'] == 'Data Engineer'].copy()

df_DE = df_DE.dropna(subset=['salary_year_avg'])

df_DE['salary_year_avg'].plot(
    kind = 'hist',
    bins = 30,
    edgecolor = 'black'
)

plt.title('Data Engineer Yearly Salaries')
plt.xlabel('Yearly Salary')
plt.ylabel('Frequency')
plt.show()
```
